```python
summary_info = {}
models_dir = "experiments/models"
model_names = listdir(models_dir)
for model_name in model_names:
    print(model_name)
    max_epoch = 100 if "lc" in model_name else 190
    train_acc = np.zeros(max_epoch//10 + 1)
    test_acc = np.zeros(max_epoch//10 + 1)
    for epoch in np.arange(0, max_epoch+1, 10):
        path = "experiments/models/" + model_name + "/epoch_" + str(epoch) + ".pt"
        data = torch.load(path, map_location='cpu')
        i = epoch // 10
        train_acc[i] = data['train_accuracy']
        test_acc[i] = data['test_accuracy']
    combined_acc = {'train': train_acc, 'test': test_acc}
    summary_info[model_name] = combined_acc
        
```

```python
import torchvision
def load_mnist():
    transform = torchvision.transforms.Compose(
        [torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((0.5,), (0.5,))])
    train_set = torchvision.datasets.MNIST("./data", train=True, download=True, transform=transform)
    test_set = torchvision.datasets.MNIST("./data", train=False, download=True, transform=transform)
    train_loader = torch.utils.data.DataLoader(train_set, batch_size=128, shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_set, batch_size=128, shuffle=False)
    return train_loader, test_loader

def load_cifar():
    transform = torchvision.transforms.Compose(
        [torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
    train_set = torchvision.datasets.CIFAR10("data", train=True, download=True, transform=transform)
    test_set = torchvision.datasets.CIFAR10("data", train=False, download=True, transform=transform)
    train_loader = torch.utils.data.DataLoader(train_set, batch_size=128, shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_set, batch_size=128, shuffle=False)
    return train_loader, test_loader

train_loader, test_loader = load_mnist()
for mnist_images, mnist_labels in train_loader:
    break
mnist_input = vnn.expand_input_conv(mnist_images, 10)

train_loader, test_loader = load_cifar()
for cifar_images, cifar_labels in train_loader:
    break
    
import torch.nn as nn
#cifar_input = vnn.expand_input(nn.Flatten()(cifar_images), 10)

cifar_input = vnn.expand_input_conv(cifar_images, 10)
```
